Завдання 5. Визначення популярності Python

```python
# Перевіряємо чи є Python
worked_with_python = df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)

# Рахуємо кількість
python_count = worked_with_python.sum()
total = df.shape[0]
percentage = (python_count / total) * 100

print(f"Відсоток респондентів: {round(percentage, 1)}%")
```

Завдання 6. Аналіз шляхів навчання програмуванню

```python
# Перенвіряємо чи навчався через онлайн-курси
learned_online_courses = df['LearnCode'].str.contains('Online Courses or Certification', case=False, na=False)

# Рахуємо кількість
online_courses_count = learned_online_courses.sum()

print(online_courses_count)
```

Завдання 7. Географічний аналіз компенсації Python-розробників

```python
#  Вибираємо тих хто працює з Python
df_python = df[df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)]

#  Видаляємо рядки без ЗП
df_python = df_python.dropna(subset=['ConvertedCompYearly'])

#  Групуємо
result = df_python.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median', 'count']).reset_index()

# Прибираємо країни з малою вибіркою
result = result[result['count'] >= 50]
result = result.drop(columns=['count'])

# Сортуємо по медіані
result = result.sort_values(by='median', ascending=False)

# Топ 10 країн по медіані компенсації
result.head(10)
```

Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів
